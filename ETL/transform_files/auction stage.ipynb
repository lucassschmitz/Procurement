### Find the date of the 'Licitaciones' for each FA code. 

```python
base_path = os.path.join(root_path, 'raw_data','Licitaciones')

years = range(2017, 2023) 
months = range(1, 13) #13
relevant_periods = []
for year in years: 
    for month in months: 
        path = os.path.join(base_path, f'{year}', f'lic_{year}-{month}.csv')
        df = pd.read_csv(path, sep = ';', encoding='latin1')

        df17 = df[df['CodigoExterno'] == '2239-4-LR17']
        df21 = df[df['CodigoExterno'] == '2239-5-LR21']
        df23 = df[df['CodigoExterno'] == '2239-8-LR23']

        rows_2017 = len(df17) 
        rows_2021 = len(df21)
        rows_2023 = len(df23) 

        if rows_2017 + rows_2021 + rows_2023 > 0:
            print(f'For {year}-{month} the number of observations are:', rows_2017, rows_2021, rows_2023)
            relevant_periods.append((year, month))

```

Merge in the reverse order, we merge auction_21 to the Adjudication df. 

```python
# Merge auction_21 and auction_21_adjudicacion on 'exact_merge_var'
merged_df = auction_21_adjudicacion.merge(auction_21, on='exact_merge_var', how='left', indicator=True)

# 1. Calculate the share of observations in auction_21 that are merged
share_merged = merged_df['_merge'].value_counts(normalize=True).get('both', 0)
print(f"Share of observations in auction_21 that are merged: {share_merged:.2%}")

# 2. Create a variable that counts how many matches are found in the other file
match_counts = auction_21.groupby('exact_merge_var').size()
auction_21_adjudicacion['match_count'] = auction_21_adjudicacion['exact_merge_var'].map(match_counts).fillna(0).astype(int)

#unmatched_df = merged_df[merged_df['_merge'] == 'left_only']
#print(unmatched_df.shape)

#show(merged_df)

```

For observations with various matches we keep the best match

```python
merged_df['Vehículo_y'] = merged_df['Marca_y'].astype(str) + ' '  + merged_df['Modelo'].astype(str) + ' '  + merged_df['Versión'].astype(str)

# 1. Create a similarity measure between 'Vehículo' and 'Vehículo_y'
def calculate_similarity(a, b):
    return SequenceMatcher(None, a, b).ratio()

merged_df['similarity'] = merged_df.apply(
    lambda row: calculate_similarity(row['Vehículo'], row['Vehículo_y']), axis=1
)

# 2. For observations where match_count_x > 1, keep only the observation with the highest similarity
merged_df = merged_df.loc[
    merged_df.groupby('Index')['similarity'].idxmax()
]
show(merged_df)
```

```python
# check the the merge is perfect
print(auction_21_adjudicacion.shape, merged_df.shape)   
```
