Transacciones CM 2023

```python
transac_cm_2023_path = os.path.join('..', 'yearly_data', 'Transacciones', 'transacciones_cm_2023.csv')
transac_cm_2023_df = pd.read_csv(transac_cm_2023_path)
transac_cm_2023_df.drop(columns=['Unnamed: 0.1', 'Unnamed: 0'], inplace=True)
transac_cm_2023_df
```

```python
transac_cm_2023_df['Fecha Envío OC'] = pd.to_datetime(transac_cm_2023_df['Fecha Envío OC'])
```

```python
transac_cm_2023_df['Fecha Envío OC YM'] = transac_cm_2023_df['Fecha Envío OC'].dt.strftime('%Y-%m')
```

```python
fig, ax = plt.subplots()
oc_fecha_cantidad = transac_cm_2023_df[['Fecha Envío OC', 'Cantidad']].groupby(['Fecha Envío OC']).count().reset_index()
plt.bar(oc_fecha_cantidad['Fecha Envío OC'], oc_fecha_cantidad['Cantidad'])
plt.title('Fecha v/s Cantidad de Órdenes de Compra')
ax.set_xlabel("Fecha")
ax.set_ylabel("Cantidad de Órdenes de Compra")
```

```python
fig, ax = plt.subplots()
plt.bar(transac_cm_2023_df['Fecha Envío OC'], transac_cm_2023_df['Cantidad'])
plt.title('Fecha envío Orden de Compra v/s Cantidad de vehículos adquiridos')
ax.set_xlabel("Fecha de Envío Orden de Compra")
ax.set_ylabel("Cantidad de vehículos adquiridos")
```

```python
fig, ax = plt.subplots()
year_month_sales = transac_cm_2023_df[['Fecha Envío OC YM', 'Cantidad']].groupby('Fecha Envío OC YM').sum().reset_index()
ax.bar(year_month_sales['Fecha Envío OC YM'], year_month_sales['Cantidad'])
plt.title('Mes de Envío Orden de Compra v/s Cantidad de vehículos')
ax.set_xlabel("Mes de Envío Orden de Compra")
ax.set_ylabel("Cantidad de vehículos")
```

```python
def parse_scientific_notation(s):
    s = str(s)
    s = s.replace(',', '.')
    if 'e' not in s:
        return s
    coefficient, exponent = s.split('e+')
    coefficient = float(coefficient.replace(',', '.'))
    exponent = int(exponent)
    value = coefficient * (10 ** exponent)
    return str(value)
```

```python
transac_cm_2023_df['Precio Unitario'] = transac_cm_2023_df['Precio Unitario'].apply(parse_scientific_notation)
transac_cm_2023_df['Precio Unitario'] = transac_cm_2023_df['Precio Unitario'].astype('float')
```

```python
transac_cm_2023_df['Precio Unitario'].info()
```

```python
sns.boxplot(x=transac_cm_2023_df['Tipo de Producto'], y=transac_cm_2023_df['Precio Unitario'])
plt.xticks(rotation=90)
plt.title('Tipo de Vehículo v/s Precio Unitario')
```

```python
for category in transac_cm_2023_df['Tipo de Producto'].unique():
    fig, ax = plt.subplots(figsize=(15,6))
    data = transac_cm_2023_df[transac_cm_2023_df['Tipo de Producto'] == category]
    plt.hist(data=data, x='Precio Unitario', bins=100, stacked=True, density=True)
    sns.kdeplot(data['Precio Unitario'])
    plt.title(f'Distribución de precios según categoría: {category}')

```
