## Licitaciones files

### Find the date of the 'Licitaciones' for each FA code. 

```python
base_path = os.path.join('..','raw_data','Licitaciones')

years = range(2022, 2025) 
months = range(1, 13) #13
relevant_periods = []
for year in years: 
    for month in months: 
        path = os.path.join(base_path, f'{year}', f'lic_{year}-{month}.csv')
        df = pd.read_csv(path, sep = ';', encoding='latin1')

        df17 = df[df['CodigoExterno'] == '2239-4-LR17']
        df21 = df[df['CodigoExterno'] == '2239-5-LR21']
        df23 = df[df['CodigoExterno'] == '2239-8-LR23']

        rows_2017 = len(df17) 
        rows_2021 = len(df21)
        rows_2023 = len(df23) 

        if rows_2017 + rows_2021 + rows_2023 > 0:
            print(f'For {year}-{month} the number of observations are:', rows_2017, rows_2021, rows_2023)
            relevant_periods.append((year, month))

```

```python
print(relevant_periods)
```

From the previous code we know that the auction for  '2239-4-LR17' is in 2017-5, for '2239-5-LR21' is in 2021-4 and for '2239-8-LR23' is in ------

```python
path17 = os.path.join(base_path, '2017', 'lic_2017-5.csv')
path21 = os.path.join(base_path, '2021', 'lic_2021-4.csv')
path23 = os.path.join(base_path, '2023', 'lic_2023-6.csv')


df  = pd.read_csv(path17, sep = ';', encoding='latin1')
df17 =  df[df['CodigoExterno'] == '2239-4-LR17']

df  = pd.read_csv(path21, sep = ';', encoding='latin1')
df21 = df[df['CodigoExterno'] == '2239-5-LR21']

df  = pd.read_csv(path23, sep = ';', encoding='latin1')
df23 = df[df['CodigoExterno'] == '2239-8-LR23']

```

```python
print("\nOne per line:")
for col in df17.columns:
    print(col)
```

```python
# If your dataframe is a Pandas DataFrame:
columns_to_keep = [
    'Link', 'CodigoExterno',
    'FechaInicio', 'FechaFinal',
    'FechaAdjudicacion', 'MontoEstimado',
    'TiempoDuracionContrato', 'NumeroOferentes',
    'Codigoitem', 'Nombre producto genrico',
    'UnidadMedida', 'Cantidad',
    'CodigoProveedor', 'CodigoSucursalProveedor',
    'RutProveedor', 'NombreProveedor',
    'Monto Estimado Adjudicado', 'Estado Oferta',
    'Cantidad Ofertada', 'Moneda de la Oferta',
    'Oferta seleccionada'
]

df17 = df17[columns_to_keep]
df21 = df21[columns_to_keep]
df23 = df23[columns_to_keep]

print(len(df17), len(df21), len(df23))


df_combined = pd.concat([df17, df21, df23], ignore_index=True)
print(df_combined.shape)
```
