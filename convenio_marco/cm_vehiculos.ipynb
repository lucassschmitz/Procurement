Datos transacciones CM 2023

Hay registros de transacciones desde enero 2024 para CM 2023 vehículos

```python
year = 2024
df_transacciones_cm_2023 = pd.DataFrame()
destination_path = os.path.join('..', 'interm_data', 'yearly_data', 'Transacciones',  f'transacciones_cm_2023.csv')

# Ensure the directory exists
destination_dir = os.path.dirname(destination_path)
if not os.path.exists(destination_dir):
    os.makedirs(destination_dir)


for month in range(1, 12 + 1):
    try:
        path = os.path.join('..', 'raw_data', 'ConvenioMarco', 'transacciones', str(year), f'{year}-{month}.csv')
        df = pd.read_csv(path)
        rows = df[df['Nro Licitación Pública'] == '2239-8-LR23']
        if len(rows) != 0:
            df_transacciones_cm_2023 = pd.concat([df_transacciones_cm_2023, rows])
    except:
        continue
print(destination_path)


# Print the absolute path of the destination
absolute_destination_path = os.path.abspath(destination_path)
print(f"CSV will be saved to: {absolute_destination_path}")


df_transacciones_cm_2023.to_csv(destination_path)
df_transacciones_cm_2023
```
