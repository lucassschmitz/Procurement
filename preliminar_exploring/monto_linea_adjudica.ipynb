Transform columns from str to int/float. Note there's some scientific notation, so we have to process this a little bit more.

```python
print(values_columns)
```

```python
def parse_scientific_notation(s):
    s = str(s)
    s = s.replace(',', '.')
    if 'e' not in s:
        return s
    coefficient, exponent = s.split('e+')
    coefficient = float(coefficient.replace(',', '.'))
    exponent = int(exponent)
    value = coefficient * (10 ** exponent)
    return str(value)
```

```python
df_2022['MontoEstimado'] = df_2022['MontoEstimado'].apply(parse_scientific_notation)
df_2022['Monto Estimado Adjudicado'] = df_2022['Monto Estimado Adjudicado'].apply(parse_scientific_notation)
df_2022['CantidadAdjudicada'] = df_2022['CantidadAdjudicada'].apply(parse_scientific_notation)
df_2022['Valor Total Ofertado'] = df_2022['Valor Total Ofertado'].apply(parse_scientific_notation)
df_2022['MontoLineaAdjudica'] = df_2022['MontoLineaAdjudica'].apply(parse_scientific_notation)
```

```python
df_2022['MontoEstimado'] = df_2022['MontoEstimado'].astype('float')
df_2022['Monto Estimado Adjudicado'] = df_2022['Monto Estimado Adjudicado'].astype('float')
df_2022['CantidadAdjudicada'] = df_2022['CantidadAdjudicada'].astype('float')
df_2022['Valor Total Ofertado'] = df_2022['Valor Total Ofertado'].astype('float')
df_2022['MontoLineaAdjudica'] = df_2022['MontoLineaAdjudica'].astype('float')
df_2022[values_columns]
```

Se puede observar que aunque un proveedor oferte cierta cantidad de productos, el encargado de la licitación puede adjudicar más unidades de lo ofertado. Por lo tanto, 'MontoLineaAdjudica' indica CantidadAdjudicada x MontoUnitarioOferta.
